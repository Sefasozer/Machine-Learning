```python
#Datadaki boş değerleri atıyoruz onun yerine ortalama değer koyuyoruz
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
yas=Data2.iloc[:,1:4].values
imputer=imputer.fit(yas[:,1:4])
yas[:,1:4]=imputer.transform(yas[:,1:4])
```

```python
#katogrikten numaric e cevirme
import sklearn as skl
from sklearn import preprocessing
```

```python
ulke=Data2.iloc[:,0:1].values
le=preprocessing.LabelEncoder()
ulke[:,0]=le.fit_transform(Data2.iloc[:,0])
```

```python
oh=preprocessing.OneHotEncoder()
ulke=oh.fit_transform(ulke).toarray()
```

```python
ulke
```

```python
#geri DataFrame olarak topluyoruz
sonuc1=pd.DataFrame(data=ulke,index=range(1,23),columns=["TR","USE","FR"])
yas1=Data2.iloc[:,1:4].values
sonuc2=pd.DataFrame(data=yas1,index=range(1,23),columns=["Boy","Kilo","Yas"])
cinsiyet=Data2.iloc[:,4].values
sonuc3=pd.DataFrame(data=cinsiyet,index=range(1,23),columns=["Cinsiyet"])
se=pd.concat([sonuc1,sonuc2],axis=1)
```

```python
se
```

```python
#Datayı train ve test data olarak bölüyoruz...
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(se,sonuc3,test_size=0.33,random_state=0)
```

```python
#Veriler 0 ile 1 arasında skale ediyoruz...
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(x_train)
X_test=sc.fit_transform(x_test)
```

```python
#verileri makineye ogretiyoruz
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression(random_state=0)
LR.fit(X_train,y_train)
Tahmin=LR.predict(X_test)
```
