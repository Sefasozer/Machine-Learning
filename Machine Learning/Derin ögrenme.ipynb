```python
#Kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Veriyi çekiyoruz
Data1=pd.read_excel("derin.xlsx")
Data1
```

```python
#Öğrenecek kolonları ve sonuç kolonlarini ayiriyoruz.
X=Data1.iloc[:,3:13].values
Y=Data1.iloc[:,13].values
```

```python
#Katagorikten Numaric e götürüyoruz
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
le1=preprocessing.LabelEncoder()
X[:,1]=le1.fit_transform(X[:,1])
le2=preprocessing.LabelEncoder()
X[:,2]=le2.fit_transform(X[:,2])
```

```python
ohe=ColumnTransformer([("ohe",OneHotEncoder(dtype=float),[1])],remainder="passthrough")
```

```python
X=ohe.fit_transform(X)
```

```python
X[0:1]
```

```python
#Verileri train ve test datası olarak ayırıyoruz.
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=0)
```

```python
#Verileri 0 ile 1 arasında skaler ediyoruz
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(x_train)
X_test=sc.fit_transform(x_test)
```

```python
#Model inşaası
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
Classifier=Sequential()
Classifier.add(Dense(6,activation="relu"))
Classifier.add(Dense(6,activation="relu"))
Classifier.add(Dense(1,activation="sigmoid"))
Classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics="accuracy")
Classifier.fit(X_train,y_train,epochs=300)
y_pred=Classifier.predict(X_test)
```

```python
#Sonucları confusion matrix ile kontrol ediyoruz
from sklearn.metrics import confusion_matrix
y_pred=(y_pred>0.5)
cm=confusion_matrix(y_test,y_pred)
```
