```python
#Katagprikten numarik e cevirmek için ayırıyoruz
outlook=Data1.iloc[:,0:1].values
orta=Data1.iloc[:,1:3].values
windy=Data1.iloc[:,3:4].values
play=Data1.iloc[:,4:5].values
```

```python
#Katagorikten numarik e ceviriyoruz
from sklearn import preprocessing
```

```python
lab=preprocessing.LabelEncoder()
outlook[:,0]=lab.fit_transform(outlook[:,0])
windy[:,0]=lab.fit_transform(windy[:,0])
play[:,0]=lab.fit_transform(play[:,0])
```

```python
one=preprocessing.OneHotEncoder()
outlook=one.fit_transform(outlook).toarray()
windy=one.fit_transform(windy).toarray()
play=one.fit_transform(play).toarray()
```

```python
play=play[:,0:1]
windy=windy[:,0:1]
```

```python
#Verileri önce dataframe yapıp sonra birleştiriyoruz.
outlook=pd.DataFrame(data=outlook,columns=["sunny","overcast","rainy"])
orta=pd.DataFrame(data=orta,columns=["temperature","humidity"])
play=pd.DataFrame(data=play,columns=["play"])
windy=pd.DataFrame(data=windy,columns=["windy"])
```

```python
hepsi1=pd.concat([outlook,orta,windy,play],axis=1)
```

```python

nantemperature=hepsi1.drop(["temperature"],axis=1)
temperature=hepsi1[["temperature"]]
nantemperature
```
