```python
#Kütüphaneler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Veriyi cekiyoruz
Data1=pd.read_excel("musteriler.xlsx")
```

```python
X=Data1.iloc[:,3:].values
```

```python
#Bu model sınıflandırma için değil sınıflandırılmış grupların içindeki farklar
#Mesela biz erkek ve kız olarak ayirdik onların icinden sigara icenler icmeyenler gibi
#k-means++
from sklearn.cluster import KMeans
```

```python
#n_clusters ile kaç grup olduğunu buluyoruz...
kmeans=KMeans(n_clusters=3,init="k-means++",random_state=0)
kmeans.fit(X)
```

```python
#Burada n_clusters i 10 a kadar değer veriyoruz en performans lı grup sayımı buluyoruz 
#ama bu grup sayısı belli değilse uygulanır genelde eğer ayırcağınız grup sayisi belli ise gerek yok :)
sonuclar=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=0)
    kmeans.fit(X)
    sonuclar.append(kmeans.inertia_)

    
```

```python
#Sonuclara göre 4 grup ideal
plt.plot(range(1,10),sonuclar)
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,init="k-means++",random_state=0)
kmeans.fit(X)
Y_tahmin=kmeans.predict(X)
```

```python
#Grupları renklendirerek görseleştiriyoruz.
plt.scatter(X[Y_tahmin==0,0],X[Y_tahmin==0,1],s=100,c="red")
plt.scatter(X[Y_tahmin==1,0],X[Y_tahmin==1,1],s=100,c="blue")
plt.scatter(X[Y_tahmin==2,0],X[Y_tahmin==2,1],s=100,c="orange")
plt.scatter(X[Y_tahmin==3,0],X[Y_tahmin==3,1],s=100,c="black")
```

```python
#Agglomerative Clustering (Hiyeraşik bölünme)
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=4,affinity="euclidean",linkage="average")
hc.fit(X)
y_tahmin=hc.fit_predict(X)
```

```python
plt.scatter(X[y_tahmin==0,0],X[y_tahmin==0,1],s=100,c="red")
plt.scatter(X[y_tahmin==1,0],X[y_tahmin==1,1],s=100,c="blue")
plt.scatter(X[y_tahmin==2,0],X[y_tahmin==2,1],s=100,c="orange")
plt.scatter(X[y_tahmin==3,0],X[y_tahmin==3,1],s=100,c="black")
```
